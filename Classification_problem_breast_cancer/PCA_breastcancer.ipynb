###### PCA or Principal component Analysis is an unsupervised algorithm which is used for machine learning and Data analysis for better Data Visualization

###### It often found that several data columns provide similar kinds of Data which are not at all useful to the model but increase the training time, complexity and Dimensionality of the model, Data and the Problem. In Such cases, it is very difficut to visualize and understand the data. So, PCA is done to find out the most important features responsible for the major part of the model learning and for visusalization. It is primarily done for Dimensionality reduction

### 2-PCA

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
PC = pca.fit_transform(X_scaled_df)
p_Df = pd.DataFrame(data = PC
             , columns = ['principal component 1', 'principal component 2'])
```

##### It is PCA projection in two dimension

##### So we our converting our multi dimensions data into two dimensions

```python
p_Df.head()
```

```python
len(p_Df)
```

```python
Y_l=Y.iloc[:]['diagnosis'].tolist()
```

```python
joined=p_Df
```

```python
joined['Diagnosis']=Y_l
```

```python
joined.head()
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize = (10,10))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('PC 1', fontsize = 12)
ax.set_ylabel('PC 2', fontsize = 12)
ax.set_title('PCA', fontsize = 15)
diags=['M','B']
colors=['r','b']
for diag, color in zip(diags,colors):
    i = joined['Diagnosis'] == diag
    ax.scatter(joined.loc[i, 'principal component 1']
               , joined.loc[i, 'principal component 2']
               , c = color
               , s = 50)
ax.legend(diags)
ax.grid()
```

##### It works on variance

##### It is the square differences from the mean. variance=((x- mean(x))^2)/n

```python
pca.explained_variance_ratio_
```

###### Now what PCA does is, here there are 30 dimension in the data. So,here the PCA will create a 2D dimensional plane for PCA 2 then it will take projection of those 30 Dimensional points on the 2D plane. Then we will be able to find the best fitting line to classify the classes. It also helps to visualize. But during the projection we also lose a lot of information or variance

###### This loss of variance is  judged and depending on this the principle components are chosen. First a best bit 2D plane is chosen and projections are taken. Then another 2D plane is chosen orthogonal to the previous one and then again the projections are taken and so on. All the planes are created orthogonally. Then the variance lost are  compared and then the least one is chosen

###### It helps a lot in importance decision like model decisions also. Like here the logistic regression will give better result than KNN and others because it creates a straight line and here it can be easily formed

### Recursive Feature Elimination

Recursive feature elimination (RFE, Guyon et al. (2002)) is basically a backward selection of the predictors. This technique begins by building a model on the entire set of predictors and computing an importance score for each predictor. The least important predictor(s) are then removed, the model is re-built, and importance scores are computed again. In practice, the analyst specifies the number of predictor subsets to evaluate as well as each subset’s size. Therefore, the subset size is a tuning parameter for RFE. The subset size that optimizes the performance criteria is used to select the predictors based on the importance rankings. The optimal subset is then used to train the final model.

### ROC and AUC curves

Reciever Operating Characteristic or ROC curve is often utilised as a visualisation plot to measure the performance of a binary classifier. It’s not a metric of the model, per se, rather the graphical representation of the True Positive Rate (TPR) versus False Positive rate (FPR) at various classification threshold from 0 to 1.
